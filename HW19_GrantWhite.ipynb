```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 120
```

```python
def kinetics(K1, K2, U0, n):
    #make A
    A = np.array([[-K1,   0, 0],
                  [ K1, -K2, 0],
                  [  0,  K2, 0]])
    
    #initialize U
    U = np.empty((n+1, 3))
    U[0] = U0
    
    #forward Euler
    k = 8/n
    for i in range(n):
        U[i+1] = U[i] + k*A@U[i]
        
    return U
```

```python
def plot_func(U, n, title=''):
    t = np.linspace(0, 8, n+1)
    plt.title(title)
    plt.plot(t, U[:,0], label='$u_1$')
    plt.plot(t, U[:,1], label='$u_2$')
    plt.plot(t, U[:,2], label='$u_3$')
    plt.legend()
    plt.show()
```

#### (1)

```python
#solve
K1, K2 = 3, 1
n = 300
U0 = np.array([3, 4, 2])
U = kinetics(K1, K2, U0, n)

#plot
plot_func(U, n, f'K1 = {K1}, K2 = {K2}')
```
