```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
real = np.linspace(-2, 2, 100)
imag = np.linspace(-2, 2, 100)
```

```python
def func(r, i):
    z = r + i*1j
    return np.abs((2+z)/(2-z))
```

```python
plt.figure(figsize=(4,4), dpi=150)

for r in real:
    if func(r, 0) <= 1:
            plt.scatter(r, 0, c='b')
    for i in imag:
        if func(0, i) <= 1:
            plt.scatter(0, i, c='b')
        if func(r, i) <= 1:
            plt.scatter(r, i, c='b')
            
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('real', size=6)
plt.ylabel('imag', size=6)
plt.show()
```
