```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
plt.rcParams["figure.dpi"] = 150
```

```python
step = 0.01
real, imag = np.mgrid[-3:3+step:step, -3:3+step:step]
z = real + imag*1j
```

```python
plt.figure(figsize=(5,5))
root = 1 + z
plt.pcolormesh(real, imag, np.abs(root) <= 1, cmap='Blues', shading='nearest')
plt.title('Example 7.4')
plt.show()
```

```python
plt.figure(figsize=(5,5))
root = 1 / (1 - z)
plt.pcolormesh(real, imag, np.abs(root) <= 1, cmap='Blues', shading='nearest')
plt.title('Example 7.4')
plt.show()
```

```python
plt.figure(figsize=(5,5))
root = (1 + z/2) / (1 - z/2)
plt.pcolormesh(real, imag, np.abs(root) <= 1, cmap='Blues', shading='nearest')
plt.title('Example 7.6')
plt.show()
```
