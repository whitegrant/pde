```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
n = 500
theta = np.linspace(0, 2*np.pi, n)
zeta = np.exp(1j*theta)
```

```python
#2-step
rho = zeta**2 - zeta
sigma = 3/2*zeta - 1/2
AB_2 = rho / sigma

#3-step
rho = zeta**3 - zeta**2
sigma = 23/12*zeta**2 - 4/3*zeta + 5/12
AB_3 = rho / sigma

#4-step
rho = zeta**4 - zeta**3
sigma = 55/24*zeta**3 - 59/24*zeta**2 + 37/24*zeta - 3/8
AB_4 = rho / sigma
```

```python
#2-step
plt.figure(figsize=(4,4), dpi=150)
plt.title('Stability region of Adams-Bashforth 2-step method')
plt.plot([-3, 1], [0, 0], 'k', lw=.3)
plt.plot([0, 0], [-2, 2], 'k', lw=.3)
plt.plot(AB_2.real, AB_2.imag, 'tab:blue', lw=.8)
plt.fill(AB_2.real, AB_2.imag, 'tab:blue', alpha=.5)
plt.xlim(-3, 1)
plt.ylim(-2, 2)
plt.show()

#3-step
plt.figure(figsize=(4,4), dpi=150)
plt.title('Stability region of Adams-Bashforth 3-step method')
plt.plot([-3, 1], [0, 0], 'k', lw=.3)
plt.plot([0, 0], [-2, 2], 'k', lw=.3)
plt.plot(AB_3.real, AB_3.imag, 'tab:blue', lw=.8)
plt.fill(AB_3.real, AB_3.imag, 'tab:blue', alpha=.5)
plt.xlim(-3, 1)
plt.ylim(-2, 2)
plt.show()

#4-step
plt.figure(figsize=(4,4), dpi=150)
plt.title('Stability region of Adams-Bashforth 4-step method')
plt.plot([-3, 1], [0, 0], 'k', lw=.3)
plt.plot([0, 0], [-2, 2], 'k', lw=.3)
plt.plot(AB_4.real, AB_4.imag, 'tab:blue', lw=.8)
plt.fill_between(AB_4.real, AB_4.imag, where=(AB_4.real<0), color='tab:blue', alpha=.5)
plt.xlim(-3, 1)
plt.ylim(-2, 2)
plt.show()
```
